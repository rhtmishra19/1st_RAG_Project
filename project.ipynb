```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain_community.vectorstores import InMemoryVectorStore
```

```python
# Document loading

loader = PyPDFLoader(r"All_pdf\1775618625407.pdf")
doc = loader.load()
```

```python
print(len(doc))
```

```python
# Splitting

splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
doc = splitter.split_documents(doc)
len(doc)
```

```python
# embedding Model.

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
```

```python
# vector Store -- Store the embeddings on vector DB
Vector_db = InMemoryVectorStore.from_documents(
                                                documents= doc, 
                                                embedding= embeddings
                                                )
```

```python
# User Query.
query = "who is author of this pdf ?"
documents = Vector_db.similarity_search(query=query, k = 2 )
print(len(documents),  "\n" ,"\n" ,"\n" , documents[0].page_content)
```

```python
print(len(documents),  "\n" ,"\n" ,"\n" , len(documents[0].page_content))
```

```python
# Context generation of all documents

context = ""
for doc in documents:
    context = context + doc.page_content + "\n\n"
```

```python
# Combinig context + Question and passing it to LLM.

prompt = f"""
You are a helpful AI assistant.

Answer the question based only on the provided context.

Context:
{context}

Question:
{query}

Answer:
"""
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
answer = llm.invoke(prompt)
```

```python
print(answer.content)
```
